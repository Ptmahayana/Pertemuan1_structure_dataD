**NAMA MAHASISWA**

```python
mahasiswa = ["Rahul", "Abdul", "Bahlil", "Yanto", "Galer"]
x = mahasiswa
print("[List Mahasiswa]")
print(x)
```

**TAMBAH NAMA MENGGUNAKAN .append()**

```python
mahasiswa.append("Bowo")
print("[Setelah Penambahan Nama Baru]")
print(x)
```

**HAPUS NAMA HAPUS 1 NAMA MENGGUNAKAN .remove()**

```python
mahasiswa.remove("Bowo")
print("[Setelah Remove 1 Nama]")
print(x)
```
