**Array 3D berukuran 2x2x2**

```python
import numpy as np
print("Array 3D Berukuran 2x2x2")
array_3d = np.array([[[1, 2],[3, 4]],[[5, 6],[7, 8]]])
print(array_3d)
```

**Menampilkan Element Pada Index [1][0][1]**

```python
print("Menampilkan Elemnt Pada Index [1][0][1]")
print(array_3d[1,0,1])
```

**Menghitung Total Seluruh Array 3D**

```python
print("Menghitung Total Seluruh Array 3D")
total = np.sum(array_3d)
print(total)
```
